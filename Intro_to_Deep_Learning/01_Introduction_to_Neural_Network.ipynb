# Lecture I: Introduction to Neural Networks


Suppose you want to calculate the temperature of an exoplanet, and you do not know any equations to do so. So, let's think about it!

Consider a case in that you can observe and measure the temperature of several exoplanets and also the temperature of their host stars. Suppose we want to predict the exoplanet temperature. In that case, we should find the best fit for this data and have a **model** to predict exoplanet temperature as a function of host stars' temperature.


In other words, the temperature of the host stars is the input to the model, and the output is the exoplanet's temperature. This is the simplest **neural network**, which only has one neuron.


But in general, the temperature of any planet depends on how much energy it receives and stores. So, the **type of the host star** is essential, whether a cold white dwarf or a red giant star. Therefore, the host star's **temperature** and the **radius** are quantities critical to estimating the planet's temperature.

In addition to those quantities, it is essential to know the **orbital distance** of the exoplanet because if it is closer to the host star, it will receive much more energy and radiation.

On the other hand, the **type of the exoplanet** also matters. The composition of the core and the atmosphere of the exoplanet are the key features of the planet to estimate how much energy can be stored and how much of the stellar radiation can be reflected from the planet's surface. Thus, there are two other quantities here: **Albedo** and **f**, which is the fraction of the radiated energy that is absorbed by the planet.  

All in all, we have a set of inputs that are used to define the **hidden units** which in this case are host star type, exoplanet type, and orbital distance:


In general, all the inputs are connected to all hidden units, and basically, we let the **Neural Networks** find the dependency of the hidden units on the inputs.


This is the well-known image of the **Standard Neural Networks**. There are also other types of neural networks such as **Convolutional Neural Networks (CNN)**, which is the method that is mostly used in image processing, and **Recurrent Neural Networks (RNN)** which is used in analyzing sequence data such as texts or audio.
