# Lecture II: Logistic Regression

Hot-Jupiters (not Jupyter! :D) are exoplanets that are so close to their host stars with orbital distances typically less than 0.05 A.U. and also they are physically similar to Jupiter which means they are giant gaseous planets. The surface temperature of them is really high that's why we called them **hot**-Jupiters. Now, consider the case that which we collected the data of several exoplanets and their host stars. what we want to know is how many of the observed exoplanets are hot-Jupiters. Let's say that if the exoplanet was hot-Jupiter we label it by **1** and if it is not, its label is **0**. In doing so, we are performing binary classification of data i.e. our output ($y$) is $y \in \{0, 1\}$. To build our neural networks for this problem we need **Logistic Regression** which is an algorithm for binary classifications. We also need a **training set**. A training set is a subset of the data that we know about the output. In our case, consider a subset of the data in which we know whether the exoplanets are hot-Jupiters or not, and use this training set to train the neural networks.

Let's summarize what we have so far. We have data of exoplanets, such as orbital distance, radius size, mass, and so on. So for each exoplanet, we have a **vector** input ($x$), let's call the length of these vectors **n**: $x \in R^n$. 

Furthermore, for some of them, let's say for **m** number of them, we also know if they are hot-Jupiter or not. It means that our training set has m elements:
$$
\{(x^1, y^1), (x^2, y^2), ..., (x^m, y^m)\}
$$
Thus, basically, our **input matrix** in our training set is


The output of our training set is also a matrix such that
$$
Y = [y^1, y^2, ..., y^m]
$$

At this step, we use a linear regression model to connect each column in the input ($X^i$) with the corresponding output ($y^i$) in our training set. In a **linear regression model** we assume this relationship between the output and the input is linear, i.e.,

\begin{equation}
z = \omega X^i + b \tag{1}
\label{eq:Z}
\end{equation} 

where $X_1 = [x]$, $\omega \in R^n$, $b \in R$ and $T$ denotes to the transpose of the matrix.
However, we should modify this relationship in order to secure that the values of $Y$ are in the range of $[0, 1]$. Since in the end, we want to know what is the **probability** of being a hot-Jupiter, we need to modify the equation eq. \ref{eq:Z} by a **sigmoid function**. A sigmoid function converts the model's output into a probability score. One of the commonest sigmoid functions is the **logistic sigmoid function** which is defined as:

\begin{equation}
\sigma(z) = \frac{1}{1 + e^{-z}} \tag{2}
\label{eq:sigmoid}
\end{equation}

Using the sigmoid function in eq. \ref{eq:Z}, as the last layer of the neural network, we have

\begin{equation}
a = \sigma (\omega X + b) \tag{3}
\label{eq:a}
\end{equation}

The sigmoid function here is an example of the **activation function** used in deep learning. If you have more than one hidden layer it is very common that people to use other forms of activation functions such as $tanh(z)$. For further information about the activation functions look at Lecture IX.
