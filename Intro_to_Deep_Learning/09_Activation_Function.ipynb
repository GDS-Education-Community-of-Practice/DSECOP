Usually, the sigmoid function is only used for the last layer in binary classification problems. For the hidden layers, $tanh(x)$ works better since $tanh(x)$ keeps the mean value of the output almost zero. But why do we need the activation function in the hidden layers? If we do not use a non-linear activation function (such as sigmoid or $tanh$), there is no point in adding more than one layer to your neural network!

The most common activation function in deep learning is the **Rectified Linear Unit** or **ReLU** function which is:

\begin{equation}
g(z) = max(0, z)
\end{equation}

```python
ReLU = [max(xi, 0.) for xi in x]
plt.plot(x, ReLU)
plt.ylabel('ReLU(x)')
plt.grid()
```

One advantage of the ReLU function is that the slope (gradient) of the ReLU function is very different than zeros, at least for half of the data ($z > 0$), which causes the learning process to happen much faster. However, one disadvantage of the ReLU function is for $z <= 0$; the slope is zero! Thus, **Leaky Rectified Linear Unit**, or **Leaky ReLU**, is the slightly modified version of the standard ReLU function. The Leaky ReLU has a small slope for negative values instead of a flat slope, and it has the form like
\begin{equation}
g(z) = max(\beta z, z)
\end{equation}

where the slope coefficient, $\beta$ is a hyperparameter, i.e., determined before training.

```python
beta = 0.05
leaky_ReLU = [max(beta * xi, xi) for xi in x]
plt.plot(x, leaky_ReLU)
plt.ylabel('Leaky ReLU(x)')
plt.grid()
```
