By adding more nodes, and more layers, we are constructing our **deep neural network**. In general, we could have a very complicated deep neural network such as


Number of layers in a deep nueral network is number of hidden layers plus one correponign to the output layer. IF we represent the input layer $X = [x_1, x_2, ..., x_n]$ by $a^{[0]}$, the 2-step computation in each nodes of each layer obey a general form:

\begin{equation}
z_j^{[l]} = (\omega^T)^{[l]}_j a^{[l - 1]} + b_j^{[l]} \tag{1}
\end{equation}
\begin{equation}
a_j^{[l]} = g^{[l]} (z_j^{[l]}) \tag{2}
\end{equation}

where $l$ represents the number of layers, $j$ is the index of the node in each layer, and $a^{[l]}$ is the output of each layer. Based on this equation, for each layer, we can define two vectors $Z^{[l]}$ and $A^{[l]}$ so that

\begin{equation}
Z^{[l]} = [z_1^{[l]}, z_2^{[l]}, ..., z_{n_l}^{[l]}] \tag{3}
\end{equation}
\begin{equation}
A^{[l]} = [a_1^{[l]}, a_2^{[l]}, ..., a_{n_l}^{[l]}] \tag{4}
\end{equation}

where $n_l$ is the number of nodes in the layer $l$.
